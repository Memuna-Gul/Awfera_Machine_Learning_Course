## **Step 2:** Remove Duplicate Rows




```python
df=df.drop_duplicates()
df.head()
```

## **Step-3:** Handling Missing Values

```python
df.isnull().sum()
```

```python
# if age or salary is missing , fill with the middle value(median)
df['Age']=df['Age'].fillna(df['Age'].median()) #inplace means do the changes in the same place
df['Salary']=df['Salary'].fillna(df['Salary'].median())

#if name or gender is missing, fill with mode
df['Name']=df['Name'].fillna(df['Name'].mode()[0])
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])

#if is active is missing , fill it with mode
df['Is Active']=df['Is Active'].fillna(df['Is Active'].mode()[0])

#if joining date is missing , fill with the earliest date (min)
df['Joining Date']=df['Joining Date'].fillna(df['Joining Date'].min())

#if Comments is missing , fill it with unknown
df['Comments']=df['Comments'].fillna('Unknown')

print("Missing vales have been filled")

```

## **Step 4:** Normalize and Scale the Data

```python
from sklearn.preprocessing import MinMaxScaler #Scales all feature values between 0 and 1
scaler=MinMaxScaler()
df[['Age','Salary']]=scaler.fit_transform(df[['Age','Salary']])
df[['Age','Salary']].head()
```

```python
from sklearn.preprocessing import StandardScaler #Shifts your data so the mean = 0 and standard deviation = 1.
scaler=StandardScaler()
df[['Age','Salary']]=scaler.fit_transform(df[['Age','Salary']])
df[['Age','Salary']].head()
```

# Encode Categorical variables

```python
#Label Encoding

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

df['Gender']=le.fit_transform(df['Gender'])
df['Is Active']=le.fit_transform(df['Is Active'])

print(df.head())
```

# Step 6: Detecting outliers using z-score method

```python
from scipy.stats import zscore
import numpy as np
#calclate z score for age and salary
z_score=np.abs(zscore(df[["Age","Salary"]]))
#identify rows with zscore>3 (outliers)
outliers=(z_score>3).any(axis=1)
#Remove outliers from the dataset
df_no_outliers=df[~outliers]
#display dataset without outliers
df_no_outliers.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('/content/employee_data_1000.csv')
```

```python
#Diplay dataset information of loaded
print("\n Dataset Preview")
print(df.head())
print("\n Dataset column names")
print(df.columns)
```

```python
# change columns names for clarity
df.columns=['ID', 'Name', 'Age', 'Gender', 'Salary', 'Joining Date', 'Is Active',
       'Comments']
df.columns
```

```python
#select numeric columns for regression
numeric_cols=df.select_dtypes(include=['int64','float64']).columns.tolist()
numeric_cols
```

```python
numeric_cols.remove('ID')
```

```python
numeric_cols
```

```python
#Define Features X and target Y
X=df[["Age"]]
Y=df['Salary']
```

```python
#split data into training and testing data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
#create and train the linear regression model
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
#predict on test data
Y_pred=model.predict(X_test)
```

```python
#display model coefficients
print("\n Model coefficients")
print(f"Intercept: {model.intercept_}")
print(f"Coefficient: {model.coef_[0]}")
```

```python
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(Y_test,Y_pred)
print(f"Mean Squared Error: {MSE}")
```
